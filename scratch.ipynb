```python
betas = betas.drop(columns=["Unnamed: 0"])
betas
```

```python
import numpy as np

betas_mean = betas.mean(axis=0)
betas_mean_array = betas_mean.to_numpy()
```

```python
print(betas_mean_array)
```

```python
import seaborn as sns
import datetime
import matplotlib.pyplot as plt

starting_date = datetime.datetime(2023, 6, 25)
dates = [starting_date + datetime.timedelta(i) for i in range(len(betas_mean_array))]
ax = sns.lineplot(
    x=range(len(betas_mean_array)),
    y=betas_mean_array,
)
ax.set(xlabel="Days after 2023-06-25", ylabel="Beta", title="MLE Beta Mean :: Arizona")
```
